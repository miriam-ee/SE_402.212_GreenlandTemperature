```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

# CSV laden
df = pd.read_csv(r"C:\Users\leapi\OneDrive - Universität Graz\Master\2. Semester\Seminar Mountain research\SE_402.212_GreenlandTemperature\Data\Data_Processed\station_pairs_monthly.csv")
# Zeitformat umwandeln
df["time"] = pd.to_datetime(df["time"])

# Nach Zeit sortieren
df = df.sort_values("time")

# Alle Stationen holen
sites = sorted(df["site"].unique())

# Stationen mit Trendlinie
trend_sites = ["KAN", "NUK", "QAS", "SCO", "THU"]

# 2x4 Subplots erstellen (8 Plots)
fig, axes = plt.subplots(2, 4, figsize=(18, 10), sharex=True, sharey=True)

# Achsen flach machen
axes = axes.flatten()

# Jede Station einzeln plotten
for ax, site in zip(axes, sites):

    df_site = df[df["site"] == site]

    ax.plot(
        df_site["time"],
        df_site["gradient_per100m"],
        linestyle="-",
        marker="o",
        markersize=3
    )


    # Trendlinie nur für bestimmte Stationen
    if site in trend_sites:

        # Zeit numerisch machen
        x = mdates.date2num(df_site["time"])
        y = df_site["gradient_per100m"]

        # NaNs entfernen
        mask = ~np.isnan(y)
        x = x[mask]
        y = y[mask]

        # Lineare Regression
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)

        # Trendlinie plotten
        ax.plot(
            df_site["time"],
            p(mdates.date2num(df_site["time"])),
            linewidth=2.5,
            linestyle="-",
            color="darkred",
            label="Trend"
        )

    # Titel pro Plot
    ax.set_title(site)

    # Jährliche Ticks
    ax.xaxis.set_major_locator(mdates.YearLocator(1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

    # Grid
    ax.grid(True)

# Gemeinsame Labels
fig.suptitle("Entwicklung des Temperaturgradienten aller Stationen", fontsize=16)

fig.text(0.5, 0.04, "Zeit", ha="center")
fig.text(0.04, 0.5, "Gradient pro 100m", va="center", rotation="vertical")

# X-Achsen Labels drehen
for ax in axes:
    plt.sca(ax)
    plt.xticks(rotation=45)

plt.tight_layout(rect=[0.05, 0.05, 1, 0.95])
plt.show()
```
